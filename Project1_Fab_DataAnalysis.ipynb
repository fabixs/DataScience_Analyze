```python
import pandas as pd
import gdown

# File ID from the Google Drive link
file_id = '1d2IF_OurBm5CJjBdjiK18bUqz4QQqci2'

download_url = f'https://drive.google.com/uc?id={file_id}'

# Download the file
output_file = 'downloaded_file.csv'
gdown.download(download_url, output_file, quiet=False)

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(output_file)
```

**PROCESS**

```python
df.head()
```

```python
print("shape of the dataset")
print(df.shape)
print("\n columns of the dataset")
print(df.columns)
print("\n data types of the dataset")
print(df.dtypes)
print("\n index of the dataset")
print(df.index)
```

```python
print("show all the unique value in the Weather column")
print(df["Weather"].unique())
```

```python
print("show all the number of unique value in the weather column")
print(df["Weather"].nunique())
```

```python
print("show the total number of  value in each column")
print(df.count())
```

```python
print("Basic informations about the dataset")
print(df.info())
print("\n Statitic information about the dataset")
print(df.describe())
print("\n missing values")
print(df.isnull().sum())
```

```python
print("Number of unique values in each column:")
print(df.nunique())
```

```python
print("different time zones")
df["Weather"].value_counts()
```

```python
print("show the unique Wind Speed in the data")
print(df['Wind Speed_km/h'].unique())
```

```python
print('Find the number of each element in  weather column')

df.Weather.value_counts()
```

```python
print('Find the number of time where the weather is exactly clair')

df[df.Weather == "Clear"]
```

```python
print('Find the number of time where the weather is exactly clair')

df.groupby('Weather').get_group('Clear')
```

```python
print("find the number of time where the Wind Speed was exactly 4km/h")
df[df["Wind Speed_km/h"]== 4]
```

```python
print("find the number of time where the Wind Speed was exactly 4km/h")

df.groupby('Wind Speed_km/h').get_group(4)
```

```python
print('Find out all the not null value in the dataset')
df.notnull().sum()
```

```python
print('Rename the column Weather of the dataframe to Weather Condition')
df.rename(columns= {'Weather' : 'Weather Condition'}, inplace=True)
df.head(2)
```

```python
print(' the mean of \'Visibility\'')
df.Visibility_km.mean()
```

```python
print("standard deviation of the pressure in the data")
df.Press_kPa.std()
```

```python
print("Variance of the relative humidity")
df["Rel Hum_%"].var()
```

```python
print(" All instances when 'Snow' was recorded")
df[df['Weather Condition'].str.contains('Snow')]
```

```python
print("All instances when Wind Speed is above 24 and Visibility is 25")
df[(df['Wind Speed_km/h'] > 24) & (df.Visibility_km == 25)]
```

```python
print('mean value for each column against weather condition')

numeric_df = df.select_dtypes(include=['number'])
df.groupby('Weather Condition')[numeric_df.columns].mean()

```

```python
numeric_df = df.select_dtypes(include=['number'])
print("\n the minimum")
print(df.groupby('Weather Condition')[numeric_df.columns].min())
print("\n the maximum")
print(df.groupby('Weather Condition')[numeric_df.columns].max())
```

```python
df.groupby('Weather Condition').get_group('Fog')
```
