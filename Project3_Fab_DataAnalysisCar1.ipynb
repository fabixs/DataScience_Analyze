```python
import pandas as pd
import gdown
# File ID from the Google Drive link
file_id = '15TeG3lLJNSFsdkdd1gswsXwn1BNdySvh'

download_url = f'https://drive.google.com/uc?id={file_id}'

# Download the file
output_file = 'downloaded_file.csv'
gdown.download(download_url, output_file, quiet=False)

# Load the CSV file into a pandas DataFrame
data = pd.read_csv(output_file)
```

```python
numeric_cols = data.select_dtypes(include=['number'])
car = data.fillna(numeric_cols.mean())
```

```python
car.isnull().sum()
```

```python
rows_with_missing_values = car[car.isnull().any(axis=1)]
display(rows_with_missing_values)
```

```python
for col in car.columns:
    print(f"Mode of column '{col}")
    print(car[col].mode()[0])
    print("=" * 20)
```

```python
for col in car.select_dtypes(include='object').columns:
    car[col].fillna({col: car[col].mode()[0]}, inplace=True)
```
